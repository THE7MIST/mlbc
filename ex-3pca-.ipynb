```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Example: Create a sample dataset (replace this with your actual data)
data = np.array([[2.5, 2.4],
                 [0.5, 0.7],
                 [2.2, 2.9],
                 [1.9, 2.2],
                 [3.1, 3.0],
                 [2.3, 2.7],
                 [2, 1.6],
                 [1, 1.1],
                 [1.5, 1.6],
                 [1.1, 0.9]])

# Standardize the data (important for PCA)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_scaled)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Print the explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Plot the principal components with orthogonal axes
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], s=100, c='blue', edgecolor='k')

# Plot orthogonal lines (axes) showing the two principal components
plt.axhline(0, color='black', linestyle='--')  # Horizontal axis
plt.axvline(0, color='black', linestyle='--')  # Vertical axis

# Label the axes
plt.title('PCA Result (Orthogonal Axes)')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # Set equal scaling for both axes

plt.show()
```
